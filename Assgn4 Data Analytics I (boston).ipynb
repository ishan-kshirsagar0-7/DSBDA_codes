# Step 1 : Import the dataset

```python
from sklearn.datasets import load_boston

boston = load_boston()
type(boston)
```

   # Step 2 : Import the required libraries

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split as tts
```

# Step 3 : Load the dataset into a dataframe

```python
df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
df_x
```

```python
df_y = pd.DataFrame(boston.target)
df_y
```

```python
df_x.describe()
```

```python
df_y.describe()
```

# Step 4 : Initializing the Linear Regression model

```python
reg = linear_model.LinearRegression()
```

# Step 5 : Spitting the dataset

```python
x_train, x_test, y_train, y_test = tts(df_x, df_y, test_size=0.3, train_size=0.7, random_state=42)
```

# Step 6 : Training the model

```python
reg.fit(x_train, y_train)
```

```python
reg.coef_
```

# Step 7 : Predicting the output

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(y_test, y_pred))
```
